# merge

```python
import numpy as np
import pandas as pd
```

```python
employees=pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['karan','rahul','sumit','mukesh','gukesh'],
    'department':['hr','it','finance','It','hr']
})

saleries=pd.DataFrame({
    'employee_id':[1,2,3,6,7],
    'salary':[600,700,800,900,300],
    'bonus':[300,400,500,600,700]
})
```

```python
employees
```

```python
saleries
```

```python
pd.merge(employees,saleries)
```

```python
pd.merge(employees,saleries,on='employee_id')
```

```python
pd.merge(employees,saleries,on='employee_id',how='inner')
```

```python
pd.merge(employees,saleries,on='employee_id',how='outer')
```

```python
pd.merge(employees,saleries,on='employee_id',how='left')
```

```python
pd.merge(employees,saleries,on='employee_id',how='right')
```

```python
pd.concat([employees,saleries])
```

```python
pd.concat([employees,saleries],axis=1)
```
