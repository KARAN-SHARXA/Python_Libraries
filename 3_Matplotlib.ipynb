```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.plot(x,y)
plt.show()
```

```python
plt.title("simple line greaph")
plt.xlabel(" x axis")
plt.ylabel(" y asis")
plt.grid()
plt.show()
```

```python
x=[1,2,3,4]
y1=[10,20,25,30]
y2=[5,15,20,25]

plt.plot(x,y1,label="class=A")
plt.plot(x,y2,label="class=B")
plt.title("class Preformance")
plt.xlabel("Test no")
plt.ylabel("score")

plt.legend()

plt.grid(True)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[5,4,3,2,1]

plt.plot(x,y,color='c',linestyle='-.',marker='o')

plt.plot()
```

```python
x=[1,2,3,4]
y1=[10,20,30,40]
y2=[40,30,20,10]

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title("Plot 1")


plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("Plot 2")


plt.tight_layout()
plt.show()
```

```python
student=['karan','pinku','anmol']
marks=[85,90,70]

plt.bar(student,marks,color='green')
plt.title("student Mraks")
plt.xlabel("student")
plt.ylabel("marks")
plt.grid(True,axis='y')
plt.show()
```

```python
data=[22,87,5,43,56]
plt.hist(data,bins=5,color='skyblue',edgecolor='black')
plt.show()
```

```python
x=[10,20,30,40,50]
y=[15,25,35,45,55]
plt.scatter(x,y,color='red',marker='^',s=100)
plt.show()
```

```python
labels=['pyhton','java','c++','javaScript']
sizes=[40,30,15,15]
plt.pie(sizes,labels=labels,startangle=90)
plt.show()
```
