# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.set_style("darkgrid")
```

```python
sns.histplot(data=tips,x="total_bill",bins=20,kde=True)
plt.show()
```

```python
sns.kdeplot(data=tips["total_bill"],fill=True)
plt.show()
```

```python
sns.displot(data=tips,x="total_bill",kde=True)
```

```python
sns.displot(data=tips,x="total_bill",kde=True,hue="sex")
plt.show()
```

```python
sns.histplot(data=tips,x="tip")
plt.show()
```

```python
sns.kdeplot(data=tips["tip"],fill=True)
```

```python
sns.displot(data=tips, x="tip",hue="smoker",kde=True)
```

```python
sns.barplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
sns.countplot(x="day",data=tips)
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip")
plt.show()
```
